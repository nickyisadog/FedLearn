### Import Library

```python
import numpy as np
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data = pd.read_csv('BankChurners_over.csv')
```

### Stratified Spliting

```python
y = pd.DataFrame(data['CreditLevel'])
X = data.drop(columns=['CreditLevel'])
```

```python
X_1, X_temp, y_1, y_temp = train_test_split(X, y, stratify=y, test_size=5/6)
X_2, X_temp, y_2, y_temp = train_test_split(X_temp, y_temp, stratify=y_temp,test_size=4/5)
X_3, X_temp, y_3, y_temp = train_test_split(X_temp, y_temp, stratify=y_temp,test_size=3/4)
X_4, X_temp, y_4, y_temp = train_test_split(X_temp, y_temp, stratify=y_temp,test_size=2/3)
X_5, X_6, y_5, y_6 = train_test_split(X_temp, y_temp, stratify=y_temp,test_size=1/2)
```

### Output 6 csv file

```python
df1 = pd.concat([X_1,y_1], axis=1)
df2 = pd.concat([X_2,y_2], axis=1)
df3 = pd.concat([X_3,y_3], axis=1)
df4 = pd.concat([X_4,y_4], axis=1)
df5 = pd.concat([X_5,y_5], axis=1)

df_train = pd.concat([df1,df2,df3,df4,df5], axis=0)
df_test = pd.concat([X_6,y_6], axis=1)
```

```python
df1.to_csv('BankChurners_fd1.csv', index = False)
df2.to_csv('BankChurners_fd2.csv', index = False)
df3.to_csv('BankChurners_fd3.csv', index = False)
df4.to_csv('BankChurners_fd4.csv', index = False)
df5.to_csv('BankChurners_fd5.csv', index = False)
df_train.to_csv('BankChurners_fdtrain.csv', index = False)
df_test.to_csv('BankChurners_fdtest.csv', index = False)
```
